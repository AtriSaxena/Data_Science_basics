```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize']=10,6
sb.set_style('whitegrid')
```

```python
#labelling plot Functional method
x=range(1,10)
y=[1,2,3,4,0.5,4,3,2,1]
plt.bar(x,y)

plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z=[1,2,3,4,0.5]
veh_type=['bicycle','motorbike','car','van','stroller']
plt.pie(z,labels=veh_type)
plt.show()
```

```python
address="C:\Data_jupyter\mtcars.csv"
cars=pd.read_csv(address)
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

#object oriented method of plotting

mpg=cars.mpg

fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')
ax.set_title('Miles per Gallon of cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gallon')
```

```python
#Add Legend to a PLOT
#Functional Method

plt.pie(z)
plt.legend(veh_type,loc='best')
plt.show()
```

```python
#Lengend using Object oriented method
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')
ax.set_title('Miles per Gallon of cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gallon')

ax.legend(loc='best')
```
